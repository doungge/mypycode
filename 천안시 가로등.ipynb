```python
import pandas as pd
import folium

a= pd.read_csv('C:/Users/sskym/Downloads/충청남도 천안시_가로등 현황_20201215.csv',thousands =',', encoding ='euc-kr' )
cc = []
tv = []

for i in range(994):
    cc.append(a.iloc[i,3])
    tv.append(a.iloc[i,4])
    
print(cc)
print(tv)
```

```python
m = folium.Map(location = (36.6617, 127.0702), tiles = 'cartodbpositron')
```

```python
for i in range(994):
    c =cc[i], tv[i]
    
    folium.Marker(c, popup = c).add_to(m)
```

```python
m.save('test.html')
```
